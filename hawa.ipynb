```python
"""In a new Code block, write a function that takes a list of strings and returns True if one of the
strings is "coconut" (in lower case), or False otherwise. Include code outside the function that
shows that the function works correctly."""

def has_coconut(strings):
    return "coconut" in (s.lower() for s in strings)

# Test cases
fruits1 = ["apple", "COCONUT", "banana"]
fruits2 = ["kiwi", "mango", "pear"]

print(has_coconut(fruits1))  # True
print(has_coconut(fruits2))  # False

```

```python
"""In a new Code block, write a function that takes as its parameter a list of integers and returns the
maximum value in the list. Include some code outside the function that shows that the function
works correctly"""


def max_value(numbers):
    if not numbers:
        return None
    max_number = numbers[0]

    for num in numbers[1:]:
        if num > max_number:
            max_number = num

    return max_number

def min_value(numbers):
    if not numbers:
        return None
    min_number = numbers[0]

    for num in numbers[1:]:
        if num < min_number:
            min_number = num
    
    return min_number

test1 = [3, 1, 4, 1, 5, 9]
test2 = [-1, -5, 0, -2]
test3 = [42]

print("Test1:", max_value(test1), min_value(test1))  # 9 1
print("Test2:", max_value(test2), min_value(test2))  # 0 -5
print("Test3:", max_value(test3), min_value(test3))  # 42 42

```

```python
"""In a new Code block, write a function that takes a list of integers and returns True if the list is
sorted in ascending order, or False otherwise. Include code outside the function to demonstrate
correctness."""

def is_sorted_ascending(num_list):
    for i in range(len(num_list)-1):
        if num_list[i] > num_list[i + 1]:
            return False
    else:
        return True


x = [1, 2, 3, 4, 5]

y = [20, 15, 11]

z = [1]

print(is_sorted_ascending(x))
print(is_sorted_ascending(y))
print(is_sorted_ascending(z))

def is_sorted(num_list):
    return num_list == sorted(num_list)

print(is_sorted(x))
print(is_sorted(y))
print(is_sorted(z))
```

```python
"""In a new Code block, write a function that takes as its parameter a list of integers. The function
should return True if every integer in the list is between 0 and 100 inclusive, or False otherwise.
Include code outside the function that shows it works correctly."""

def in_between_1_100(*nums):
    for x in nums:
        if not 1 <= x <= 100:
            return False
    else:
        return True
    

print(in_between_1_100(1,2,3))

print(in_between_1_100(101, -1, 1))

def in_between_1_100(*nums):
    return all(1 <= x <= 100 for x in nums)


# Demonstration
print(in_between_1_100(1, 2, 3))       # True
print(in_between_1_100(101, -1, 1))    # False
```

```python
"""In a new Code block, write a function that takes a list of strings and returns True if the list
contains no empty strings, or False otherwise. Include code to demonstrate correctness.
Twentieth commit."""


# Function to check if a list contains no empty strings
def no_empty_strings(string_list):
    for s in string_list:
        if s.strip() == "":  # Treat empty or whitespace-only strings as empty
            return False
    return True


# Demonstration code
list1 = ["apple", "banana", "cherry"]
list2 = ["apple", "", "cherry"]
list3 = ["apple", "   ", "banana"]

print(no_empty_strings(list1))  # True
print(no_empty_strings(list2))  # False
print(no_empty_strings(list3))  # False
```
